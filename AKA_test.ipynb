```python
import pandas as pd
```

```python
#reading csv file
df = pd.read_csv('/content/Test CSV 1 - PS4.csv', encoding='utf8')
```

```python
# adding ps4 game console column before joining with xbox
df['Game_Console'] = 'PS4'
df.head()
```

```python
df1 = pd.read_csv('/content/Test CSV 2 - Xbox One.csv', encoding='utf8')
```

```python
# adding xbox one game console column before joining with ps4
df1['Game_Console'] = 'Xbox One'
df1.head()
```

```python
#outer joining to return all values from csv1 and csv2
#suffixes used to differentiate xbox columns from ps4 columns
DF_A = df.merge(df1, on=('Game', 'Year', 'Genre', 'Publisher'), how='outer', suffixes=('_ps4', '_xboxone'))
DF_A.head()
```

```python
df2 = pd.read_csv('/content/Test CSV 3 - All Platforms.csv', encoding='utf8')
df2.shape
```

```python
df2.head()
```

```python
#fill nan with 0 to not lose other important information
df2.fillna(0)
```

```python
#grouping by name and release year to perform multiple additions on all sales data and critic/user count, and averages on critic/user score
DF_B = df2.groupby(['Name','Year_of_Release']).agg({'NA_Sales': 'sum', 'EU_Sales' : 'sum', 'JP_Sales' : 'sum', 'Other_Sales' : 'sum', 'Global_Sales' : 'sum', 'Critic_Score' : 'mean', 'Critic_Count' : 'sum', 'User_Score' : 'mean', 'User_Count' : 'sum'})
DF_B.shape
```

```python
#groupby changed Name and Year_of_Release from columns to to indexes 
DF_B.reset_index(inplace=True) #change index to column for easier merge
DF_B.head(10)
```

```python
#inner joining to show games only in both DF_B and DF_A
DF_C = DF_A.merge(DF_B, left_on='Game', right_on='Name', suffixes=('', '_all'))
DF_C.shape
```

```python
DF_C.head()
```

```python
#POGS = percentage of global sales
DF_C['POGS_ps4'] = (DF_C['Global_ps4'] / DF_C['Global_Sales']) * 100
DF_C['POGS_xboxone'] = (DF_C['Global_xboxone'] / DF_C['Global_Sales']) * 100
DF_C.head()
```
