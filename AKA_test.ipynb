```python
import pandas as pd
```

```python
#reading csv file
df = pd.read_csv('/content/Test CSV 1 - PS4.csv', encoding='utf8')
```

```python
# adding ps4 game console column before joining with xbox
df['Game_Console'] = 'PS4'
df.head()
```

```python
df1 = pd.read_csv('/content/Test CSV 2 - Xbox One.csv', encoding='utf8')
```

```python
# adding xbox one game console column before joining with ps4
df1['Game_Console'] = 'Xbox One'
df1.head()
```

```python
#outer joining to return all values from csv1 and csv2
#suffixes used to differentiate xbox columns from ps4 columns
DF_A = df.merge(df1, on=('Game', 'Year', 'Genre', 'Publisher'), how='outer', suffixes=('_ps4', '_xboxone'))
DF_A.head()
```

```python
df2 = pd.read_csv('/content/Test CSV 3 - All Platforms.csv', encoding='utf8')
df2.shape
```

```python
df2.head()
```

```python
#fill nan with 0 to not lose other important information
df2.fillna(0)
```

```python
#grouping by name and release year to perform multiple additions on all sales data and critic/user count, and averages on critic/user score
DF_B = df2.groupby(['Name','Year_of_Release']).agg({'NA_Sales': 'sum', 'EU_Sales' : 'sum', 'JP_Sales' : 'sum', 'Other_Sales' : 'sum', 'Global_Sales' : 'sum', 'Critic_Score' : 'mean', 'Critic_Count' : 'sum', 'User_Score' : 'mean', 'User_Count' : 'sum'})
DF_B.shape
```

```python
#groupby changed Name and Year_of_Release from columns to to indexes 
DF_B.reset_index(inplace=True) #change index to column for easier merge
DF_B.head(10)
```

```python
#inner joining to show games only in both DF_B and DF_A
DF_C = DF_A.merge(DF_B, left_on=('Game', 'Year'), right_on=('Name', 'Year_of_Release'), suffixes=('', '_all'))
DF_C.shape
```

```python
DF_C.head()
```

```python
#manually adding _all suffix 
DF_C = DF_C.rename(columns={'NA_Sales': 'North America_all', 'EU_Sales': 'Europe_all', 'JP_Sales': 'Japan_all', 'Other_Sales': 'Rest of Wordld_all', 'Global_Sales': 'Global_all'})
```

```python
#POGS = percentage of global sales
DF_C['POGS_ps4'] = (DF_C['Global_ps4'] / DF_C['Global_all']) * 100
DF_C['POGS_xboxone'] = (DF_C['Global_xboxone'] / DF_C['Global_all']) * 100
DF_C.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Sorting year column in ascending order so that trend charts make sense
DF_C.sort_values(["Year"], 
                    axis=0,
                    ascending=[True], 
                    inplace=True)
```

```python
DF_C.tail()
```

```python
#selecting games only on ps4
Games_ps4_only = DF_C[(DF_C['Game_Console_ps4'] == 'PS4') & (DF_C['Game_Console_xboxone'] != 'Xbox One') ]

Games_ps4_only.shape
```

There are 168 PS4 exclusive games

```python
#adding new column for average global sales for ps4 exclusive games
Games_ps4_only['Average_Global_ps4'] = Games_ps4_only.groupby('Year').Global_ps4.transform('mean')
Games_ps4_only.head()
```

```python
#selecting games only on xbox
Games_xbox_only = DF_C[(DF_C['Game_Console_ps4'] != 'PS4') & (DF_C['Game_Console_xboxone'] == 'Xbox One') ]
Games_xbox_only.shape
```

It seems there are no xbox exclusive games

```python
#selecting games on both ps4 and xbox
Games_both = DF_C[(DF_C['Game_Console_ps4'] == 'PS4') & (DF_C['Game_Console_xboxone'] == 'Xbox One') ]
Games_both.shape
```

The rest of the games are available on xboxone and ps4

```python
#creating new column for sum of global sales for ps4 and xboxone
Games_both['Global_ps4_xboxone'] = Games_both.groupby('Game').Global_ps4.transform('sum')

#creating new column for average of global sales for ps4 and xboxone
Games_both['Average_Global_both'] = Games_both.groupby('Year').Global_ps4_xboxone.transform('mean')

Games_both.head()
```

```python
#what was the trend of average global sales for ps4 exclusive games vs games available on both ps4 and xboxone
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(Games_ps4_only["Year"], Games_ps4_only["Average_Global_ps4"], label="ps4", color="purple")
ax.plot(Games_both["Year"], Games_both["Average_Global_both"], label="both", color="blue")
plt.xlabel("Year")
plt.ylabel("Global sales")
plt.legend()
plt.title('Average global sales of games on ps4 vs ps4 and xbox')
plt.show()
```

```python
#is there a correlation between global sales for ps4 and global sales for xboxone
sns.lmplot(x="Global_ps4", y="Global_xboxone", data=DF_C); 
plt.show()
```

```python
DF_C['Global_xboxone'].corr(DF_C['Global_ps4'])
```

There appears to be a very strong positive relationship between global ps4 sales and xboxone sales

```python
#is there a correlation between global sales and critic scores?
sns.lmplot(x="Critic_Score", y="Global_all", data=DF_C); 
plt.show()
```

```python
DF_C['Critic_Score'].corr(DF_C['Global_all'])
```

There appears to be almost no relationship between critic scores and global sales

```python
#distribution of games released per year
plt.figure(figsize=(7, 5))
sns.set(style="darkgrid")
sns.histplot(data=DF_C, x="Year", color="firebrick", label="Games per year distribution",kde=True)

plt.legend() 
plt.show()
```

From the distribution, the year with the most games released is 2016

```python
#top 10 games with highest total global sales for ps4 and xboxone
highest_gs_games = Games_both.sort_values(['Global_ps4_xboxone'], ascending = False)
g=sns.barplot(x=highest_gs_games['Game'][:10],y=highest_gs_games['Global_ps4_xboxone'][:10], palette = 'viridis')
g.set_title("Games with highest global sales", weight = "bold")
plt.xticks(rotation = 90)
plt.show()
```

```python
#top 10 games with highest critic scores for ps4 and xboxone
highest_gs_games = Games_both.sort_values(['Critic_Score'], ascending = False)
g=sns.barplot(x=highest_gs_games['Game'][:10],y=highest_gs_games['Critic_Score'][:10], palette = 'viridis')
g.set_title("Games with highest critic scores", weight = "bold")
plt.xticks(rotation = 90)
plt.show()
```
